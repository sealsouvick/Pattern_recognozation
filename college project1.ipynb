```python
import random
import math
import numpy
import matplotlib.pyplot as plt

# radius of the circle
r = int(input("Enter the radious of circle: "))
# center of the circle (x, y)
circle_x = int (input("Enter the X cordinate of the Center: "))
circle_y = int (input("Enter the Y cordinate of the Center: "))
points = int(input("No points of circle: "))
a=[]
b=[]

# random angle
for i in range(points):
    rho = round(random.uniform(0,r),2)
    phi = round(random.uniform(0,2*math.pi),2)
    a.append(circle_x+math.sqrt(r*rho)*math.cos(phi))
    b.append(circle_y+math.sqrt(r*rho)*math.sin(phi))
    
new_a=[ round(e,2) for e in a]
new_b=[ round(e,2) for e in b]
#print(new_a)
#print(new_b)
plt.scatter(new_a,new_b,s=1)
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
plt.figure(figsize =(10,10))
plt.title("Scatter Diagram After using K-means Algorithm")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.scatter(df1['X'],df1['Y'],color='green',label = 'Cluster 1',s=2)
plt.scatter(df2['X'],df2['Y'],color='red',label = 'Cluster 2',s=2)
plt.scatter(df3['X'],df3['Y'],color='yellow',label = 'Cluster 3',s=2)
plt.scatter(df4['X'],df4['Y'],color='blue',label = 'Cluster 4',s=2)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],s=200, color= 'black', label = 'Centroid')   

plt.xlim([0,160])
plt.ylim([0,160])
#plt.xlim([0,max(x+side,circle_x+r,y+side,circle_y+r)+10])
#plt.ylim([0,max(x+side,circle_x+r,y+side,circle_y+r)+10])
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(df)
data_transformed = mms.transform(df)
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```
