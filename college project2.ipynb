```python
import random
import matplotlib.pyplot as plt
 
# Function to generate
# and append them
# start = starting range
# end = ending range
# num = number of
# elements needs to be appended
def Rand(start, end, num):
	res = []

	for j in range(num):
		res.append(round(random.uniform(start, end),2))

	return res
n=int(input("enter the no retaingle generated"))
list1=[]
list2=[]
list3=[]
list4=[]
# Driver Code
for i in range(0,n):
    
    x =int(input("Enter the x cordinate of point: "))
    y=int(input("Enter the y cordinate of point:  "))
    l=int(input("Enter the length  "))
    b=int(input("Enter the bredth  "))
    N=int(input("The no of points input"))
    if i== 0:
        list1.append(x)
        list2.append(y)
        list1=(Rand(x, x+l, N))
        list2=(Rand(y, y+b, N))
    if i== 1:
        list3.append(x)
        list4.append(y)
        list3=(Rand(x, x+l, N))
        list4=(Rand(y, y+b, N))
        
plt.scatter(list1 ,list2,s=2)
#print(list1)
#print(list2)

#print(list4)



plt.scatter(list3 ,list4,s=2)  
```

```python
X=[]
Y=[]
X=list1+list3
Y=list2+list4
plt.figure(figsize =(5,5))
plt.title("Scatter Diagram")
plt.xlim([0,max(y+b,x+l)+10])
plt.ylim([0,max(y+b,x+l)+10])
plt.scatter(X,Y,s=2)
```

```python
dict={'X':X,'Y':Y}
import pandas as pd
df=pd.DataFrame(dict)
df.to_csv('matrix3.csv',index=False)
```

```python
import pandas as pd


df1=pd.read_csv("matrix3.csv")
df1.head()
```

```python
import sklearn
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
km
```

```python
y_predicted=km.fit_predict(df1[['X','Y']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
plt.figure(figsize =(5,5))
plt.title("Scatter Diagram After using K-means Algorithm")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.scatter(df1['X'],df1['Y'],color='green',label = 'Cluster 1',s=2)
plt.scatter(df2['X'],df2['Y'],color='red',label = 'Cluster 2',s=2)
#plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],s=200, color= 'yellow', label = 'Centroid')   
plt.xlim([0,max(y+b,x+l)+10])
plt.ylim([0,max(y+b,x+l)+10])
#plt.legend()
```
