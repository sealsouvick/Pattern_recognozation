```python
import random
import math
import numpy
import matplotlib.pyplot as plt

# radius of the circle
r = int(input("Enter the radious of circle: "))
# center of the circle (x, y)
circle_x = int (input("Enter the X cordinate of the Center: "))
circle_y = int (input("Enter the Y cordinate of the Center: "))
points = int(input("No points of circle: "))
a=[]
b=[]

# random angle
for i in range(points):
    rho = round(random.uniform(0,r),2)
    phi = round(random.uniform(0,2*math.pi),2)
    a.append(circle_x+math.sqrt(r*rho)*math.cos(phi))
    b.append(circle_y+math.sqrt(r*rho)*math.sin(phi))
    
new_a=[ round(e,2) for e in a]
new_b=[ round(e,2) for e in b]
#print(new_a)
#print(new_b)
plt.scatter(new_a,new_b,s=1)


```

```python
r1 = int(input("Enter the radious of circle: "))
# center of the circle (x, y)
circle_x1 = int (input("Enter the X cordinate of the Center: "))
circle_y1= int (input("Enter the Y cordinate of the Center: "))
points = int(input("No points of circle: "))
a1=[]
b1=[]

# random angle
for i in range(points):
    rho = round(random.uniform(0,r1),2)
    phi = round(random.uniform(0,2*math.pi),2)
    a1.append(circle_x1+math.sqrt(r1*rho)*math.cos(phi))
    b1.append(circle_y1+math.sqrt(r1*rho)*math.sin(phi))
    
new_a1=[ round(e,2) for e in a1]
new_b1=[ round(e,2) for e in b1]
#print(new_a1)
#print(new_b1)
plt.scatter(new_a1,new_b1,s=4)

```

```python
X=[]
Y=[]
X=new_a+new_a1
Y=new_b+new_b1
plt.figure(figsize =(7,7))
plt.title("Scatter Diagram")
plt.xlim([0,150])
plt.ylim([0,150])
plt.scatter(X,Y,s=2)
```

```python
dict={'X':X,'Y':Y}
import pandas as pd
df=pd.DataFrame(dict)
df.to_csv('matrix4.csv',index=False)
```

```python
df1=pd.read_csv("matrix4.csv")
df1.head()
```

```python
import sklearn
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
km
```

```python
y_predicted=km.fit_predict(df[['X','Y']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
plt.figure(figsize =(5,5))

plt.xlim([0,150])
plt.ylim([0,150])
plt.title("Scatter Diagram After using K-means Algorithm")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.scatter(df1['X'],df1['Y'],color='green',label = 'Cluster 1',s=2)
plt.scatter(df2['X'],df2['Y'],color='red',label = 'Cluster 2',s=2)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],s=100, color= 'yellow', label = 'Centroid')  
plt.legend()
```
