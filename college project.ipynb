```python
# Python code to generate
# random numbers and 
# append them to a list 

import random
import matplotlib.pyplot as plt
 
# Function to generate
# and append them
# start = starting range
# end = ending range
# num = number of
# elements needs to be appended
def Rand(start, end, num):
	res = []

	for j in range(num):
		res.append(round(random.uniform(start, end),2))

	return res

# Driver Code
x =int(input("Enter the x cordinate of point: "))
y=int(input("Enter the y cordinate of point:  "))
side=int(input("Enter the side of the square:  "))
N=int(input("The no of points input"))
list1=[]
list2=[]
list1.append(x)
list2.append(y)
list1=Rand(x, x+side, N)
list2 =Rand(y, y+side, N)
#print(list1)
#print(list2)

plt.scatter(list1 ,list2)
  

   
 
 
# close the file

```

```python
import random
import math
import numpy
import matplotlib.pyplot as plt

# radius of the circle
r = int(input("Enter the radious of circle: "))
# center of the circle (x, y)
circle_x = int (input("Enter the X cordinate of the Center: "))
circle_y = int (input("Enter the Y cordinate of the Center: "))
points = int(input("No points of circle: "))
a=[]
b=[]

# random angle
for i in range(points):
    rho = round(random.uniform(0,r),2)
    phi = round(random.uniform(0,2*math.pi),2)
    a.append(circle_x+math.sqrt(r*rho)*math.cos(phi))
    b.append(circle_y+math.sqrt(r*rho)*math.sin(phi))
    
new_a=[ round(e,2) for e in a]
new_b=[ round(e,2) for e in b]
#print(new_a)
#print(new_b)
plt.scatter(new_a,new_b)
```

```python


X=[]
Y=[]
X=new_a+list1
Y=new_b+list2
plt.figure(figsize =(7,7))
plt.title("Scatter Diagram")

plt.xlim([0,max(circle_x+r,x+side,circle_y+r,y+side)+20])
plt.ylim([0,max(circle_y+r,y+side,circle_x+r,x+side)+20])
plt.scatter(X,Y,s=1)


```

```python


dict={'X':X,'Y':Y}
import pandas as pd
df=pd.DataFrame(dict)
df.to_csv('matrix1.csv',index=False)
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
plt.figure(figsize =(6,6))
plt.title("Scatter Diagram After using K-means Algorithm")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.scatter(df1['X'],df1['Y'],color='green',label = 'Cluster 1',s=4)
plt.scatter(df2['X'],df2['Y'],color='red',label = 'Cluster 2',s=4)
#plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],s=100, color= 'yellow', label = 'Centroid')   
plt.xlim([0,max(circle_x+r,x+side,circle_y+r,y+side)+10])
plt.ylim([0,max(circle_y+r,y+side,circle_x+r,x+side)+10])
plt.legend()
#plt.scatter(X,Y,s=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(df)
data_transformed = mms.transform(df)
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```
